```python
with open("in2.txt", 'r') as f:
    data = f.read().splitlines()
print(data)
values_1 = []
values_2 = []
for item in data:
    item = item.strip()
    items = item.split(" ")
    values_1.append(int(items[0]))
    values_2.append(int(items[-1]))
print(values_1)
print(values_2)
```

```python
val_1 = sorted(values_1)
val_2 = sorted(values_2)
print(val_1)
print(val_2)

s = 0
for idx, value in enumerate(zip(val_1, val_2)):
    a,b = value
    # print(a,b)
    diff = a - b
    s += abs(diff)
print(s)
```

```python
s = 0
counts = {}

for value in values_1:
    if value not in counts.keys():
        if value in values_1:
            x = sum([1 if value == x else 0 for x in values_2])
            counts[value] = x
        else: counts[value] = 0

    s += value * counts[value]

print(s)
```
