```python
#convert the board
board = []
for row in range(len(data)):
    r = []
    for col in range(len(data[row])):
        if data[row][col] == '.':
            r.append(-1)
        else:
            r.append(int(data[row][col]))
    board.append(r)
print(board)
```

```python
from collections import deque

def valid_cords(cords, board):
    x, y  = cords
    if x >= 0 and x < len(board) and y >= 0 and y < len(board[0]):
        return True
    return False

def count_dfs_paths(start, board):
    stack = deque([start])
    ends = 0
    visited = set()

    while len(stack) > 0:
        node = stack.pop()
        x, y, val  = node
        visited.add((x,y))
        next_val = val + 1
        next_coords = ([x+1, y], [x-1, y], [x, y+1], [x, y-1])
        if val == 9:
            ends += 1
        for a,b in next_coords:
            if valid_cords((a,b), board) and board[a][b] == next_val and (a,b) not in visited:
                stack.append((a,b, next_val))
    return ends
```

```python
res = 0
print(board)
for row in range(len(data)):
    for col in range(len(data[row])):
        if board[row][col] == 0:
            count = count_dfs_paths((row, col, 0), board)
            print(count)
            res += count
print(res)
```

```python
def count_bfs_paths(start, board):
    queue = deque([start])
    ends = 0
    while len(queue) > 0:
        node = queue.popleft()
        x, y, val  = node
        next_val = val + 1
        next_coords = ([x+1, y], [x-1, y], [x, y+1], [x, y-1])
        if val == 9:
            ends += 1
        for a,b in next_coords:
            if valid_cords((a,b), board) and board[a][b] == next_val:
                queue.append((a,b, next_val))
    return ends
```

```python
res = 0
print(board)
for row in range(len(data)):
    for col in range(len(data[row])):
        if board[row][col] == 0:
            count = count_bfs_paths((row, col, 0), board)
            res += count
print(res)
```
