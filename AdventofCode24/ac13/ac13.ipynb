```python
import re

reg = r'Button ([A|B]): X[+](\d*), Y[+](\d*)'
reg_prize = r'Prize: X[=](\d*), Y[=](\d*)'

res = []
for i in range (len(data)):
    if re.match(reg, data[i]):
        res.append(re.match(reg, data[i]).groups())
    if re.match(reg_prize, data[i]):
        res.append(re.match(reg_prize, data[i]).groups())
print(res)
```

```python
items = []
i = 0
while i < len(res):
    x = []
    for j in range(i,i+3):
        x.append(res[j])
    items.append(x)
    i+=3
items
```

```python
import numpy as np

A = np.array([[int(res[0][1]), int(res[1][1])],[int(res[0][2]), int(res[1][2])]])
print(A)

b = np.array([int(res[2][0]), int(res[2][1])])
print(b)

x = np.linalg.solve(A, b)
x
```

```python
import math as Math

results = []
the_rest= []
for item in items:
    A = np.array([[int(item[0][1]), int(item[1][1])],[int(item[0][2]), int(item[1][2])]])
    b = np.array([int(item[2][0]) + 10000000000000, int(item[2][1]) + 10000000000000])
    x = np.linalg.solve(A, b)
    # print(A)
    # print(b)
    # print(x)
    # print(np.round(x[0]))
    # check if x is integer
    res1 = np.round(x[0]) * A[0][0] + np.round(x[1]) * A[0][1]
    res2 = np.round(x[0]) * A[1][0] + np.round(x[1]) * A[1][1]
    # print(b)
    # print(res1, res2)
    # print(sum(b == [res1, res2]))
    if sum(b == [res1, res2]) == 2:
        print("SOLVABLE")
        print(item)
        results.append(x)
    else:
        # print("WRONG")
        # print(x)
        # print(b)
        # print(res1, res2)
        the_rest.append(x)

results
# the_rest
```
