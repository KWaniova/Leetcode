```python
H, W = 71,71
blocked = blocked[:1024]
for i in range(H):
    for j in range(W):
        if (j,i) in blocked:
            print("#", end="")
        else:
            print(".", end="")
    print()
```

```python
def find_min_path(H,W,blocked):
    def dfs(x,y):
        if x == H-1 and y == W-1:
            return 0
        if (x,y) in visited:
            return float('inf')
        visited.add((x,y))
        if x < H-1 and (x+1,y) not in blocked:
            right = dfs(x+1,y)
        else:
            right = float('inf')
        if y < W-1 and (x,y+1) not in blocked:
            down = dfs(x,y+1)
        else:
            down = float('inf')
        if x > 0 and (x-1,y) not in blocked:
            left = dfs(x-1,y)
        else:
            left = float('inf')
        if y > 0 and (x,y-1) not in blocked:
            up = dfs(x,y-1)
        else:
            up = float('inf')
        visited.remove((x,y))
        return 1 + min(right,down,left,up)
    visited = set()
    return dfs(0,0)
print(find_min_path(H,W,blocked))
```
