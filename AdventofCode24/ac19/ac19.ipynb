```python
def create_word_dict(word):
    d = defaultdict(set)
    for i in range(len(word)):
        for inst in instructions:
            if word[i:].startswith(inst):
                if i + len(inst) <= len(word):
                    d[i].add(inst)

    return d


res = []
# data = ["bwurrg"]

for word in data:
    print(word)
    # print(len(word))
    possible_dic = create_word_dict(word)
    # print(possible_dic)
    # res.append(possible)
    keys = list(possible_dic.keys())
    res_dict = defaultdict(int)
    for k in keys[::-1]:
        for inst in possible_dic[k]:
            # print("k >> ", k)
            # print("inst >> ", inst)
            if k+len(inst) - 1 <= len(word):
                if k + len(inst) == len(word):
                    res_dict[k] += 1
                elif k + len(inst) in possible_dic:
                    if res_dict[k+len(inst)] > 0:
                        res_dict[k] += res_dict[k+len(inst)]

    # print(res_dict[0])

    # print(res_dict)
    s = 0
    t = ""
    for k in keys:
        text = ",".join(list(possible_dic[k]))
        t = t + " | " + " >> ".join(["i: ", str(k), "v: ", str(res_dict[k]), text])
        s+=1
        if s == 3:
            print(t)
            s = 0
            t = ""
    print(t)

    res.append(res_dict[0])



print(res)
print(res[-3:])


print(sum(res))
```
