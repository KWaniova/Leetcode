```python
# with open("instructions.txt", 'r') as f:
#     instructions = f.read()
# print(instructions)

with open("int.txt", 'r') as f:
    data = f.read().splitlines()
print(data)
print(len(data))

start, end = (0,0), (0,0)
board = []
for i in range(len(data)):
    row = []
    for j in range(len(data[i])):
        row.append(data[i][j])
        if data[i][j] == "S":
            start = (i,j)
        if data[i][j] == "E":
            end = (i,j)
    board.append(row)

# print(board)
```

```python
start, end
```

```python
directions = [(-1,0), (0,1), (1,0), (0,-1)]

def show_board(board):
    for row in board:
        print(" ".join(row))
    print("\n")

def is_valid(board, x, y):
    return 0 <= x < len(board) and 0 <= y < len(board[0])

def dfs(board, start):
    path = set([])
    value = 0
    position = start
    while board[position[0]][position[1]] != "E":
        value += 1
        for d in directions:
            next_x, next_y = position[0] + d[0], position[1] + d[1]
            if board[next_x][next_y] == ".":
                position = (next_x, next_y)
                board[next_x][next_y] = str(value)
                path.add((next_x, next_y))
                break
            if board[next_x][next_y] == "E":
                position = (next_x, next_y)
                break
    return path, value

path, length = dfs(board, start)
```

```python
show_board(board)
```

```python
res = []
for item in path:
    print(item)
    x,y = item
    value = int(board[x][y])
    for direction in directions:
        next_x, next_y = item[0] + direction[0], item[1] + direction[1]
        if is_valid(board, next_x, next_y) and board[next_x][next_y] == "#":
            nextn_x, nextn_y = next_x + direction[0], next_y + direction[1]
            if is_valid(board, nextn_x, nextn_y) and board[nextn_x][nextn_y] not in ["#", "S"]:
                if board[nextn_x][nextn_y] == "E":
                    print("found")
                    res.append((length - value -2, next_x, next_y, ))
                else:
                    try:
                        val = int(board[nextn_x][nextn_y])
                    except:
                        val = 0
                        print("error")
                    if val > value + 1:
                        res.append((val - value - 2, next_x, next_y))

# print(res)
res.sort()
# print(res[::-1])
r = [1 if x[0] >=100 else 0 for x in res ]
# print(len(res))
print(sum(r))
```
