```python
with open("in.txt", 'r') as f:
    data = f.read()
print(data)

with open("in.txt", 'r') as f:
    data = f.read().splitlines()
print(data)
```

```python
import re
#detect two overlaping strings in python with group
reg = r'(?=(XMAS|SAMX))'
horizontal = []
for row in data:
    # print(row)
    h = re.findall(reg, row)
    # print(h)
    if h:
        horizontal = [*horizontal, *h]
horizontal
```

```python
# print(data)
transposed = []
for i in range(len(data[0])):
    transposed.append(''.join([row[i] for row in data]))
transposed
```

```python
vertical = []
for col in transposed:
    h = re.findall(reg, col)
    if h:
        vertical = [*vertical, *h]
vertical
```

```python
diagonal_data_r = set([])
for i in range(len(data)):
    d = "".join(data[j][i+j] for j in range(len(data)) if 0 <= i+j < len(data))
    diagonal_data_r.add(d)
    if i > 0:
        d = "".join(data[i+j][j] for j in range(len(data)) if 0 <= i+j < len(data))
        diagonal_data_r.add(d)
        # print(d)

# print(diagonal_data_r)
diagonal_r = []
for row in diagonal_data_r:
    h = re.findall(reg, row)
    if h:
        diagonal_r = [*diagonal_r, *h]
# print(diagonal_r)
```

```python
diagonal_data_l = set([])
for i in range(len(data)):
    d = "".join(data[j][i-j] for j in range(len(data)) if 0 <= i-j < len(data))
    diagonal_data_l.add(d)
    # print(d)
    if i > 0:
        d = "".join(data[i+j][len(data) - 1 - j] for j in range(len(data)) if 0 <= i+j < len(data))
        diagonal_data_l.add(d)
        # print(d)

# print(diagonal_data_l)
diagonal_l = []
for row in diagonal_data_l:
    h = re.findall(reg, row)
    if h:
        diagonal_l = [*diagonal_l, *h]
# print(diagonal_l)
```

```python
result = horizontal + vertical + diagonal_r + diagonal_l
len(result)
```

```python
reg = r'(MAS|SAM)'
xmas_count = 0
for i in range(len(data) - 2):
    row = data[i]
    for j in range(len(row)):
        corner_idx = [(i, j), (i, j+2), (i+2, j), (i+2, j+2)]
        for idxs in corner_idx:
            if idxs[0] >= len(data) and idxs[1] >= len(row):
                continue
        x_1 = "".join([data[i+count][j + count] for count in range(3) if 0 <= j + count < len(data)])
        x_2 = "".join([data[i+2 - count][j + count] for count in range(3) if 0 <= j + count < len(data)])
        # print(x_1)
        # print(x_2)
        if re.match(reg, x_1) and re.match(reg, x_2):
            xmas_count += 1
xmas_count
```
