```python
def get_anthenas_by_row(grid):
    anthenas = {}
    for row in range(len(grid)):
        for col in range(len(grid[row])):
            val = grid[row][col]
            if val not in [".", "#"]:
                if row not in anthenas.keys():
                    anthenas[row] = [(col, val)]
                else:
                    anthenas[row].append((col, val))
    return anthenas

anthenas = get_anthenas_by_row(data)
print(anthenas)
anthenas_rows = anthenas.keys()
anthenas_rows = sorted(anthenas_rows)
print(anthenas_rows)
```
