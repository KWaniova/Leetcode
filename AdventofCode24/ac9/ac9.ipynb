```python
def convert_to_board(line):
    is_file = True
    idx = 0
    res = ""
    for c in line:
        val = int(c)
        if is_file and val > 0:
            part = [f'{idx}'] * val
            print(part)
            res += "".join(part)
            idx += 1
        elif val > 0:
            part = ["."] * val
            res += "".join(part)
        is_file = not is_file
    return res

board = convert_to_board(data)
print(board)
print(board[0:50])
print(board[len(board) - 50:])
```
