```python
import cv2
import dlib

# Load dlib's pre-trained face detector and facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

# Start video capture
cap = cv2.VideoCapture(0)

# Function to calculate the midpoint between two points
def midpoint(p1, p2):
    return (p1.x + p2.x) // 2, (p1.y + p2.y) // 2

# Function to detect gaze direction and provide feedback
def detect_gaze_and_feedback(eye_center, face_center):
    diff_x = eye_center[0] - face_center[0]
    diff_y = eye_center[1] - face_center[1]

    # Tolerance level for considering gaze as centered
    tolerance_x = 15
    tolerance_y = 10

    if abs(diff_x) < tolerance_x and abs(diff_y) < tolerance_y:
        return "Looking at the camera"
    else:
        if diff_x > tolerance_x:
            print("Look left")
        elif diff_x < -tolerance_x:
            print("Look right")

        if diff_y > tolerance_y:
            print("Look up")
        elif diff_y < -tolerance_y:
            print("Look down")

# Main loop
while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces using dlib
    faces = detector(gray)
    for face in faces:
        # Get the landmarks for the face
        landmarks = predictor(gray, face)

        # Get coordinates of left and right eye
        left_eye = landmarks.part(36)  # Left eye start point
        right_eye = landmarks.part(45)  # Right eye end point

        # Calculate the mid-point of the eyes
        eye_center = midpoint(left_eye, right_eye)

        # Calculate the center of the face
        face_center = ((face.left() + face.right()) // 2, (face.top() + face.bottom()) // 2)

        # Draw landmarks and lines
        cv2.circle(frame, eye_center, 5, (0, 255, 0), -1)  # Mark the eye center
        cv2.circle(frame, face_center, 5, (255, 0, 0), -1)  # Mark face center

        # Analyze gaze direction and give feedback
        gaze_feedback = detect_gaze_and_feedback(eye_center, face_center)
        print(gaze_feedback)

    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
